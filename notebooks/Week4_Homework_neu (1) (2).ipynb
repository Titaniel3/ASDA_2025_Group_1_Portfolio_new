```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from scipy.stats import skew, shapiro
```

```python
url = "https://docs.google.com/spreadsheets/d/1ecopK6oyyb4d_7-QLrCr8YlgFrCetHU7-VQfnYej7JY/export?format=xlsx"
dataset = pd.ExcelFile(url, engine='openpyxl')

sheets = []
for sheet in dataset.sheet_names:
    df = dataset.parse(sheet)
    df["City"] = sheet #adding a column to track from which group is the data
    sheets.append(df)

df = pd.concat(sheets, ignore_index=True)
```

# Advanced Cleaning

The Step: "Instead of having two columns of room_shared and room_private, create one with the respective categories." is not clear to me. There is already the variable room_type, which kind of makes room_shared and rooom_private redundant. The research paper states that those are dummy variables. Why cant we just drop them both?

Delete the dummy variables for the number of listings of each host:

```python
# create new colomn "host_portfolio"
# "one" if multi and biz are both 0
df["host_portfolio"] = "one"

# If multi is 1, the host has two to four listings
df.loc[df["multi"] == 1, "host_portfolio"] = "two_to_four"

# If biz is 1, the host has more than four listings
df.loc[df["biz"] == 1, "host_portfolio"] = "more_than_four"

# test if it worked
print(df["host_portfolio"].value_counts())
```

```python
# delete dummys
df = df.drop(columns=["multi", "biz"])
print(df.columns)
```

```python
# Plot the distribution of price (former realSum) for both weekday and weekend offers
df.groupby('days')['Price'].plot(kind='kde', legend=True)
plt.show()

# Check normality, skewness, modality

print("Skewness:", skew(df['Price']))
print("Shapiro-Wilk p-value:", shapiro(df['Price'].sample(5000, random_state=1)).pvalue)

# If skewed, consider using transformations (e.g. log); can you normalize it?
df['log_Price'] = np.log1p(df['Price'])  # log1p = log(1 + x) to avoid log(0)
print("Skewness after log:", skew(df['log_Price']))
df['log_Price'].plot(kind='kde', title='Log-Transformed Price Distribution')
plt.show()
```

```python
# Difference in price per night between two cities

weekdays = df[df['days'] == 'weekdays']
city_pairs = [('london','paris'), ('amsterdam','berlin'), ('rome','athens')]
for c1, c2 in city_pairs:
    prices1 = weekdays[weekdays['City'] == c1]['Price']
    prices2 = weekdays[weekdays['City'] == c2]['Price']

    if prices1.empty or prices2.empty:
        print(f"Skipping {c1} vs {c2} (no data)")
        continue

    t_stat, p_val = stats.ttest_ind(prices1, prices2)

    print(f"{c1} vs {c2} | mean1: {prices1.mean():.2f}, mean2: {prices2.mean():.2f}, p-value: {p_val:.4f}")
    print("Difference is significant" if p_val < 0.05 else "Difference is NOT significant")
```

```python

import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

# --- If df isn't loaded yet, uncomment: ---
# df = pd.read_csv("airbnb_cleaned.csv")

# 1) Clean up and standardize room_type labels (optional but helpful)
rt = (
    df["room_type"]
    .astype(str)
    .str.strip()
    .str.lower()
    .map({
        "entire home/apt": "Entire home/apt",
        "entire home/apartment": "Entire home/apt",
        "private room": "Private room",
        "shared room": "Shared room"
    })
)
df = df.assign(room_type=rt)

# 2) Define a consistent order of categories for the stack
room_order = ["Entire home/apt", "Private room", "Shared room"]

# 3) Compute relative frequencies per City (rows sum to 1)
counts = (
    df.groupby(["City", "room_type"])
      .size()
      .unstack(fill_value=0)
      .reindex(columns=room_order, fill_value=0)
)
rel = counts.div(counts.sum(axis=1), axis=0)

# 4) Plot stacked relative bar chart
ax = rel.plot(
    kind="bar",
    stacked=True,
    figsize=(12, 7),
    color=["#4C78A8", "#F58518", "#54A24B"],  # one color per room_type
    edgecolor="white",
)


# 5) Formatting
ax.set_title("Share of Room Types by City", fontsize=14)
ax.set_xlabel("City", fontsize=12)
ax.set_ylabel("Share", fontsize=12)
ax.yaxis.set_major_formatter(PercentFormatter(1.0))
ax.set_ylim(0, 1)
plt.xticks(rotation=30, ha="right")

# 6) Move legend outside of the plot area
ax.legend(
    title="Room Type",
    bbox_to_anchor=(1.05, 1),   # move legend outside the plot
    loc="upper left",
    frameon=False
)

plt.tight_layout(rect=[0, 0, 0.85, 1])  # leave space on the right for the legend
plt.show()
```

```python

import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

# --- If df isn't loaded yet, uncomment: ---
# df = pd.read_csv("airbnb_cleaned.csv")

# 1) Define order of categories (top-to-bottom stack order)
portfolio_order = ["one", "two_to_four", "more_than_four"]

# 2) Compute relative frequencies per City
counts = (
    df.groupby(["City", "host_portfolio"])
      .size()
      .unstack(fill_value=0)
      .reindex(columns=portfolio_order, fill_value=0)
)
rel = counts.div(counts.sum(axis=1), axis=0)

# 3) Plot stacked relative bar chart
fig, ax = plt.subplots(figsize=(12, 7))
rel.plot(
    kind="bar",
    stacked=True,
    ax=ax,
    color=["#4C78A8", "#F58518", "#54A24B"],  # consistent colors
    edgecolor="white"
)

# 4) Formatting
ax.set_title("Share of Host Portfolio by City", fontsize=14)
ax.set_xlabel("City", fontsize=12)
ax.set_ylabel("Share", fontsize=12)
ax.yaxis.set_major_formatter(PercentFormatter(1.0))
ax.set_ylim(0, 1)
plt.xticks(rotation=30, ha="right")

# 5) Legend outside the plot
ax.legend(
    title="Host Portfolio",
    bbox_to_anchor=(1.05, 1),
    loc="upper left",
    frameon=False
)

plt.tight_layout(rect=[0, 0, 0.85, 1])  # space for legend
plt.show()
```
